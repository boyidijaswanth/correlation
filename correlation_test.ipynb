```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
mdata = pd.read_csv("dataset.csv")
```

```python
relation = mdata.corr()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(relation,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(mdata.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(mdata.columns)
ax.set_yticklabels(mdata.columns)
plt.show()
```
